```python
import pandas as pd
import numpy as np
from matplotlib import pyplot
import itertools
import warnings
from statsmodels.tsa.stattools import acf, pacf
import math
import datetime
```

```python
mydata = pd.read_csv("train.csv")
```

```python
mydata.head(5)
```

```python
mydata.columns
```

```python
subset = mydata.loc[mydata.building_number == 1]
```

```python
subset = subset[['timestamp', 'main_meter']]
subset['timestamp'] = pd.to_datetime(subset['timestamp'], format = '%Y-%m-%d %H:%M:%S')
subset = subset.sort_values(by=['timestamp'])
```

```python
subset = subset.loc[subset.timestamp >= '2017-09-01 00:00:00']
subset.shape
```

```python
explore_df = subset.loc[subset.timestamp >= '2017-12-25 00:00:00']
explore_df = explore_df.loc[explore_df.timestamp <= '2017-12-27 00:00:00']
#explore_df
```

```python
t1 = [datetime.datetime.time(d) for d in subset['timestamp']]
t = [tm.hour for tm in t1]
exo = [1 if hr>=6 & hr<=18 else 0 for hr in t]
exo = np.array(exo)
```

```python
subset = subset.set_index('timestamp')
subset = subset[['main_meter']]
```

```python
import matplotlib.pylab as plt
lag_acf = acf(subset, nlags=5)
lag_pacf = pacf(subset, nlags=5, method='ols')

#Plot ACF: 
plt.subplot(121) 
plt.plot(lag_acf)
plt.axhline(y=0,linestyle='--',color='gray')
plt.axhline(y=-1.96/np.sqrt(len(subset)),linestyle='--',color='gray')
plt.axhline(y=1.96/np.sqrt(len(subset)),linestyle='--',color='gray')
plt.title('Autocorrelation Function')

#Plot PACF:
plt.subplot(122)
plt.plot(lag_pacf)
plt.axhline(y=0,linestyle='--',color='gray')
plt.axhline(y=-1.96/np.sqrt(len(subset)),linestyle='--',color='gray')
plt.axhline(y=1.96/np.sqrt(len(subset)),linestyle='--',color='gray')
plt.title('Partial Autocorrelation Function')
plt.tight_layout()
```

```python
train = subset.iloc[0:26304]
test = subset.iloc[26304:26401]
```

```python
train = subset.iloc[0:11616]
test = subset.iloc[11616:11712]
exo_train = exo[0:11616]
exo_test = exo[11616:11712]
```

```python
p=d=q=range(0,3)
P=D=Q=range(0,2)
pdq = list(itertools.product(p,d,q))

seasonal_pdq = [(x[0], x[1], x[2], 96) for x in list(itertools.product(P,D,Q))]
```

```python
from statsmodels.tsa.arima_model import ARIMA
```

```python
warnings.filterwarnings("ignore")
for param in pdq:
    try:
        mod = ARIMA(train, order=param)
        results = mod.fit(disp=0)
        print(param, results.aic)
    except:
        continue
```

```python
mod = ARIMA(train, order=(2,0,2))
r = mod.fit(disp=0)
r.forecast()
r.aic
```

```python
r.forecast(10)[0]
```

```python
RMSE = math.sqrt(sum((r.forecast(96)[0] - np.array(test.main_meter))**2))
RMSE
```

```python
MAPE = 100*(sum(abs((r.forecast(96)[0] - np.array(test.main_meter))/np.array(test.main_meter)))/96)
MAPE
```

```python
import statsmodels.api as sm
```

```python
warnings.filterwarnings("ignore")
pdq = [(2,0,2)]
for param in pdq:
    for param_seasonal in seasonal_pdq:
        try:
            mod = sm.tsa.statespace.SARIMAX(train, order=param, seasonal_order=param_seasonal,
                                                     enforce_stationarity=False, enforce_invertibility=False)
            results = mod.fit()
            print('ARIMA{}x{}96 - AIC:{}'.format(param, param_seasonal, results.aic))
        except:
            continue

```

```python
warnings.filterwarnings("ignore")
for param in pdq:
    for param_seasonal in seasonal_pdq:
        try:
            mod = sm.tsa.statespace.SARIMAX(train, order=param, seasonal_order=param_seasonal,
                                                     enforce_stationarity=False, enforce_invertibility=False)
            results = mod.fit()
            print('ARIMA{}x{}96 - AIC:{}'.format(param, param_seasonal, results.aic))
        except:
            continue
```

```python
mod = sm.tsa.statespace.SARIMAX(train.main_meter, exog = exo_train, order=(2, 0, 2),seasonal_order=(0,1,0,96), 
                                enforce_stationarity=False, enforce_invertibility=False)
r = mod.fit()
r.aic
```

```python
RMSE = math.sqrt(sum((np.array(r.predict(start = '2017-12-31 00:00:00', end = '2017-12-31 23:45:00', exog = pd.DataFrame(exo_test)))-np.array(test.main_meter))**2))
RMSE
```
